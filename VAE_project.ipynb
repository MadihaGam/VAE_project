```python
%matplotlib inline
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

# Gaussian VAE
The resulls displayed here are the latent space from the Gaussian VAE along with the reconstructions. Latent space dimension is 3, but by changing latent_dim variable, reults for other latent dimensions can be displayed. It is also possible to display different reconstruction epochs by changing the epoch variable.

These results are obtained from running the python script GVAE.py.

```python
latent_dim = 3
fig_folder = Path(f"/mnt/c/Users/madiha/Documents/school/Variationa_Autoencoder_Project/final_code/GVAE/latent_dim_{latent_dim}")

# Load all PNG images
pngs = sorted(fig_folder.glob("*.png"))
imgs = [Image.open(p) for p in pngs]

# latent space traversal images
plt.subplot(1, 4, 1)
plt.imshow(imgs[0])
plt.title(pngs[0].stem)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(imgs[1])
plt.title(pngs[1].stem)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(imgs[2])
plt.title(pngs[2].stem)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(imgs[3])
plt.title(pngs[3].stem)
plt.axis('off')

#figure size
plt.gcf().set_size_inches(25, 10)

plt.show()

# Reconstructed image at a specific epoch
epoch = 50
recon_image = fig_folder / "recon_images" / f"recon_test_epoch_{epoch}.png"
recon_img = Image.open(recon_image)
plt.figure()
plt.imshow(recon_img)
plt.axis('off')
plt.show()


```
